```python
import os
import matplotlib.pyplot as plt
```

```python
base_dir = '/home/roderickmajoor/Desktop/Master/Thesis/GT_data/'

# Initialize variables to store counts
subdir_counts = {}
total_count = 0

# Traverse through each subdirectory
for subdir in os.listdir(base_dir):
    subdir_path = os.path.join(base_dir, subdir)
    if os.path.isdir(subdir_path):
        # Count jpg files in the current subdirectory
        jpg_count = sum(1 for file in os.listdir(subdir_path) if file.lower().endswith('.jpg'))
        subdir_counts[subdir] = jpg_count
        total_count += jpg_count
```

```python
plt.figure(figsize=(10, 6))
plt.bar(subdir_counts.keys(), subdir_counts.values(), color='skyblue')
plt.xlabel('Subsets')
plt.ylabel('Number of Pages')
plt.title('Number of Pages in Each Subset')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Print total count
print(f'Total number of JPG files in all subdirectories: {total_count}')
```
